```python
#Look at certain column 
df['Cabin']
```

```python
# Filtering
# Access a group of rows and columns by label(s) or a boolean array.
FareAbove20_OlderThan40=df.loc[(df['Age'] > 40) & (df['Fare'] > 20)]
FareAbove20_OlderThan40
```

```python
#Find mean of a column
df['Age'].mean()
```

```python
#Fillna
df['Cabin'] = df['Cabin'].fillna('Top of the Ship')
df['Cabin']
```

```python
#Dummy Encoding 
df['male']=0
df['female']=0
df
```

```python
#Fill in columns for dummy variables correctly
for x in range(0, len(df)):
    if df['Sex'][x] == 'male':
        df['male'][x] = 1
    else:
        df['female'][x] = 1

df
```
